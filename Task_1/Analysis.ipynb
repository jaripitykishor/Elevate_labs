```python
import pandas as pd

# Load dataset
df = pd.read_csv("Medical Appointment No Shows.csv")
df.head()
```

```python
# 1. Handle missing values
print("Missing values before:\n", df.isnull().sum())
df = df.fillna("Unknown")   # simple fill for now
df.head()
```

```python
# 2. Remove duplicate rows
df = df.drop_duplicates()
df.head()
```

```python
# 3. Standardize column headers
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
df
```

```python
# 4. Standardize text values
if "gender" in df.columns:
    df["gender"] = df["gender"].astype(str).str.upper().replace({"M": "Male", "F": "Female"})
df
```

```python
# 5. Convert date columns to datetime
for col in ["scheduledday", "appointmentday", "scheduled_day", "appointment_day"]:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")
df
```

```python
# 6. Convert no_show column (Yes/No -> 1/0)
if "no_show" in df.columns:
    df["no_show"] = df["no_show"].astype(str).str.strip().str.upper()
    df["no_show"] = df["no_show"].replace({"YES": 1, "NO": 0, "UNKNOWN": None})
    df["no_show"] = pd.to_numeric(df["no_show"], errors="coerce").fillna(0).astype(int)
df
```

```python
 #7. Ensure numeric columns are correct type (example: age)
if "age" in df.columns:
    df["age"] = pd.to_numeric(df["age"], errors="coerce")
    # Fix invalid ages (<0 or >120)
    df.loc[(df["age"] < 0) | (df["age"] > 120), "age"] = df["age"].median()
    df["age"] = df["age"].astype(int)
df
```

```python
# 8. Clean neighbourhood column
if "neighbourhood" in df.columns:
    df["neighbourhood"] = df["neighbourhood"].astype(str).str.strip().str.title()

# Save cleaned dataset
df.to_csv("Medical_Appointment_NoShows_Cleaned.csv", index=False)

print("✅ Cleaning done. Final shape:", df.shape)
df
```
